## Deliverable 2. Create a Customer Travel Destinations Map.

```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps
import requests
import gmaps.datasets

# Import API key
from configGmap import gkey

# Configure gmaps API key
gmaps.configure(gkey)
```

```python
# 1. Import the WeatherPy_database.csv file. / 5. In Step 1 import the WeatherPy...
city_data_df = pd.read_csv("weather_data/WeatherPy_Database.csv")
city_data_df.head()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria 
# 6. In Step 2, write two input statements that prompt...
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.
# In Step 3, use the loc method to filter the city_data_df DataFrame for temperature criteria...
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & (city_data_df["Max Temp"] >= min_temp)]
preferred_cities_df.head()
```

```python
# 4a. Determine if there are any empty rows.
# In Steps 4a-b, determine if there are any empty rows, then drop them if necessary and create a new DataFrame.
preferred_cities_df.isnull().sum()
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesn’t have empty rows.
# In Steps 4a-b, determine if there are any empty rows, then drop them if necessary and create a new DataFrame.
clean_df = preferred_cities_df
clean_df
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
# In Steps 5a-b, use the provided code to create a new DataFrame, hotel_df...
hotel_df = clean_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()

# 5b. Create a new column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# 6a. Set parameters to search for hotels with 5000 meters.
# 10. In Step 6a, we have supplied the search parameters, which are the same as in this module...
params = {
    "radius": 5000,
    "type": "lodging",
    "key": gkey
}

# 6b. Iterate through the hotel DataFrame.
# 11. In Steps 6b-f, iterate through the hotel_df DataFrame, retrieve the latitude and longitude of each city...
for index, row in hotel_df.iterrows():
    
    # 6c. Get latitude and longitude from DataFrame
    lat = row["Lat"]
    lng = row["Lng"]
    params["location"] = f"{lat},{lng}"
    
    # 6d. Set up the base URL for the Google Directions API to get JSON data.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # 6e. Make request and retrieve the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
   
    
    # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError, ConnectionError):
        print("Hotel not found... skipping.")
        
```

```python
# 7. Drop the rows where there is no Hotel Name.
# In Step 7, drop any rows in the hotel_df DataFrame where a hotel name is not found.
# hotel_df.isnull().sum() - Shows no nulls in the data.
hotel_df.head()
```

```python
# 13.In Steps 8a-b, create an output file to store the hotel_df DataFrame
# as WeatherPy_vacation.csv in the Vacation_Search folder.
clean_hotel_df = hotel_df
#8a. Create the output File (CSV)
output_data_file = "weather_data/WeatherPy_vacation.csv"

# 8b. Export the City_Data into a csv
clean_hotel_df.to_csv(output_data_file, index_label="City_ID")
```

```python
# 14. In Step 9, add the city name, the country code, the weather...
# 9. Using the template add city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Description</dt><dd>{Max Temp} °F and {Current Description}.</dd>
</dl>
"""

# 15. In Step 10a, use the provided list comprehension code to retrieve the city data...
# 10a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in clean_hotel_df.iterrows()]

# 16. In Step 10b, use the provided code snippet to retrieve the latitude and longitude ...
# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = clean_hotel_df[["Lat", "Lng"]]
```

```python
#  17. In Steps 11a-b, refactor your previous marker layer map code to create a marker layer map...
# 11a. Add a marker layer for each city to the map. 
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# 11b. Display the figure
fig
```
